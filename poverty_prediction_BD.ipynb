```python
# Test 3: Get mean elevation for a small area
image = ee.Image('USGS/SRTMGL1_003')
point = ee.Geometry.Point([90.4125, 23.8103])  # Dhaka, Bangladesh
region = point.buffer(10000)  # 10km buffer

mean_elev = image.reduceRegion(
    reducer=ee.Reducer.mean(),
    geometry=region,
    scale=30
).getInfo()

print("Mean elevation around Dhaka (m):", mean_elev)
```
