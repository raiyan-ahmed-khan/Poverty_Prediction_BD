```python
import osmnx as ox
import geopandas as gpd
import pandas as pd

# Load your clean shapefile
districts = gpd.read_file("D:/Projects/Poverty Predictor Bd/data/raw/shapefiles/bangladesh_districts_clean.gpkg")
print(f"Loaded {len(districts)} districts")
```

```python
# Extract road network length per district
# WARNING: This takes 20-40 minutes total — let it run
road_data = []

for idx, row in districts.iterrows():
    district = row['district_name']
    try:
        G = ox.graph_from_polygon(row.geometry, network_type='drive')
        total_length_km = sum(d['length'] for u, v, d in G.edges(data=True)) / 1000
        road_data.append({
            'district_name': district,
            'division_name': row['division_name'],
            'road_length_km': round(total_length_km, 2)
        })
        print(f"✓ {district}: {total_length_km:.1f} km")
    except Exception as e:
        road_data.append({
            'district_name': district,
            'division_name': row['division_name'],
            'road_length_km': None
        })
        print(f"✗ {district}: {e}")

road_df = pd.DataFrame(road_data)
road_df.to_csv("data/raw/road_network.csv", index=False)
print(f"\nDone! Saved {len(road_df)} rows")
print(f"Missing: {road_df['road_length_km'].isnull().sum()}")
```
